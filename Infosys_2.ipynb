```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = '/content/drive/MyDrive/Project of Infosys/IPL_Data.csv'
df = pd.read_csv(file_path)
df
```

```python
# Display basic information
df.info()
df.head()
```

```python
numeric_cols = [
    "MatchPlayed", "InningsBatted", "NotOuts", "RunsScored", "100s", "50s", "4s", "6s",
    "BattingAVG", "BattingS/R", "CatchesTaken", "StumpingsMade", "Ducks", "R/O",
    "InningsBowled", "Overs", "Maidens", "RunsConceded", "Wickets", "3s", "5s",
    "BowlingAVG", "EconomyRate", "S/R", "Mtc"]
```

```python
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Drop unnecessary columns
df.drop(columns=["Sport"], inplace=True, errors='ignore')
```

```python
# Fill missing numeric values with 0 (assuming missing means no performance recorded)
df[numeric_cols] = df[numeric_cols].fillna(0)
```

```python

# Set visualization style
sns.set_style("darkgrid")
```

```python
# Plot Total Runs Scored Per Over
plt.figure(figsize=(8, 5))
sns.histplot(df["RunsScored"], bins=20, kde=False, color="yellow")
plt.title("Distribution of Total Runs Scored Per Player")
plt.xlabel("Runs Scored")
plt.ylabel("Count")
plt.show()
```

```python
# Top 10 Batsmen by Strike Rate
top_batsmen = df.nlargest(10, "BattingS/R")[["Name", "BattingS/R"]]
plt.figure(figsize=(10, 5))
sns.barplot(y=top_batsmen["Name"], x=top_batsmen["BattingS/R"], palette="viridis")
plt.title("Top 10 Batsmen by Strike Rate")
plt.xlabel("Strike Rate")
plt.ylabel("Player")
plt.show()
```

```python
# Most Sixes Hit by Players
top_six_hitters = df.nlargest(10, "6s")[["Name", "6s"]]
plt.figure(figsize=(10, 5))
sns.barplot(y=top_six_hitters["Name"], x=top_six_hitters["6s"], palette="viridis")
plt.title("Most Sixes Hit by Players")
plt.xlabel("Total Sixes")
plt.ylabel("Player")
plt.show()
```

```python

# Top 10 Most Economical Bowlers
top_economy_bowlers = df[df["EconomyRate"] > 0].nsmallest(10, "EconomyRate")[["Name", "EconomyRate"]]
plt.figure(figsize=(10, 5))
sns.barplot(y=top_economy_bowlers["Name"], x=top_economy_bowlers["EconomyRate"], palette="crest")
plt.title("Top 10 Most Economical Bowlers")
plt.xlabel("Economy Rate")
plt.ylabel("Player")
plt.show()
```

```python

# Wicket Distribution Across Overs
plt.figure(figsize=(8, 5))
sns.histplot(df["Wickets"], bins=15, kde=False, color="red")
plt.title("Distribution of Wickets Taken Per Player")
plt.xlabel("Wickets Taken")
plt.ylabel("Count")
plt.show()
```

```python

# Dot Ball Percentage by Bowler
df["Dot Ball %"] = ((df["Maidens"] * 6) / (df["Overs"] * 6)) * 100  # Approximation
dot_ball_bowlers = df.nlargest(10, "Dot Ball %")[["Name", "Dot Ball %"]]
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(y=dot_ball_bowlers["Name"], x=dot_ball_bowlers["Dot Ball %"], palette="deep")
plt.title("Top 10 Bowlers by Dot Ball Percentage")
plt.xlabel("Dot Ball Percentage")
plt.ylabel("Player")
plt.show()
```

```python

# Total Runs Scored in Powerplay & Death Overs
df["Powerplay Runs"] = df["RunsScored"] * 0.3  # Approximate assumption
df["Death Over Runs"] = df["RunsScored"] * 0.2
```

```python

plt.figure(figsize=(8, 5))
sns.histplot(df["Powerplay Runs"], bins=15, color="green", label="Powerplay Runs")
sns.histplot(df["Death Over Runs"], bins=15,  color="purple", label="Death Over Runs")
plt.title("Runs Scored in Powerplay & Death Overs")
plt.xlabel("Runs")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
# Distribution of Runs Per Ball (1s, 2s, 4s, 6s)
runs_distribution = df[["4s", "6s"]].copy()
runs_distribution["1s"] = df["RunsScored"] - (df["4s"] * 4 + df["6s"] * 6)
runs_distribution["2s"] = runs_distribution["1s"] // 2
runs_distribution["1s"] = runs_distribution["1s"] % 2

plt.figure(figsize=(8, 5))
runs_distribution.mean().plot(kind="bar", color=["blue", "green", "red", "purple"])
plt.title("Distribution of Runs Per Ball (1s, 2s, 4s, 6s)")
plt.xlabel("Run Type")
plt.ylabel("Average Count Per Player")
plt.show()
```

```python

import plotly.express as px

# Assuming 'df' is your DataFrame

if "Dismissal Type" in df.columns:
    fig = px.histogram(df, y="Dismissal Type",
                     title="Distribution of Dismissal Types",
                     labels={"Dismissal Type": "Dismissal Type", "count": "Count"},  # Customize axis labels
                     color_discrete_sequence=px.colors.qualitative.Pastel)  # Use Pastel colors

    fig.update_layout(bargap=0.2)  # Adjust bar spacing for better visibility
    fig.show()
```

```python
# Average Runs Per Over
df["Runs Per Over"] = df["RunsScored"] / df["Overs"]
df["Runs Per Over"].fillna(0, inplace=True)

plt.figure(figsize=(8, 5))
sns.histplot(df["Runs Per Over"], bins=20,  color="orange")
plt.title("Distribution of Average Runs Per Over")
plt.xlabel("Runs Per Over")
plt.ylabel("Count")
plt.show()
```

```python

# Top 10 Bowlers Conceding Most Boundaries
df["Total Boundaries Conceded"] = df["4s"] + df["6s"]
top_boundary_bowlers = df.nlargest(10, "Total Boundaries Conceded")[["Name", "Total Boundaries Conceded"]]

plt.figure(figsize=(10, 5))
sns.barplot(y=top_boundary_bowlers["Name"], x=top_boundary_bowlers["Total Boundaries Conceded"], palette="Reds")
plt.title("Top 10 Bowlers Conceding Most Boundaries")
plt.xlabel("Total Boundaries Conceded")
plt.ylabel("Bowler")
plt.show()
```

```python
# Total Extra Runs Conceded by Teams
if "Extras" in df.columns:
    total_extras = df.groupby("Team")["Extras"].sum().reset_index()
    plt.figure(figsize=(10, 5))
    sns.barplot(y=total_extras["Team"], x=total_extras["Extras"], palette="Purples_r")
    plt.title("Total Extra Runs Conceded by Teams")
    plt.xlabel("Total Extra Runs")
    plt.ylabel("Team")
    plt.show()
```

```python
# Top 10 Batsmen by Strike Rate
top_batsmen = df.nlargest(10, "BattingS/R")[["Name", "BattingS/R"]]
plt.figure(figsize=(10, 5))
sns.barplot(y=top_batsmen["Name"], x=top_batsmen["BattingS/R"], palette="viridis")
plt.title("Top 10 Batsmen by Strike Rate")
plt.xlabel("Strike Rate")
plt.ylabel("Player")
plt.show()
```

```python
# Wicket Distribution Across Overs
plt.figure(figsize=(8, 5))
sns.histplot(df["Wickets"], bins=15, color="red")
plt.title("Distribution of Wickets Taken Per Player")
plt.xlabel("Wickets Taken")
plt.ylabel("Count")
plt.show()
```
