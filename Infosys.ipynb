```python
df_IPL = pd.read_csv('/content/drive/MyDrive/Project of Infosys/IPL_Data.csv')
df_IPL
```

```python
df_IPL.info()
```

```python
numeric_cols = [
    "MatchPlayed", "InningsBatted", "NotOuts", "RunsScored", "100s", "50s", "4s", "6s",
    "BattingAVG", "BattingS/R", "CatchesTaken", "StumpingsMade", "Ducks", "R/O",
    "InningsBowled", "Overs", "Maidens", "RunsConceded", "Wickets", "3s", "5s",
    "BowlingAVG", "EconomyRate", "S/R", "Mtc"]
```

```python
# Converted all values to numeric
df_IPL[numeric_cols] = df_IPL[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Drop unnecessary columns only written (IPL and Cricket)
df_IPL.drop(columns=["Sport"], inplace=True, errors='ignore')
```

```python
# Fill missing numeric values with 0 (assuming missing means no performance recorded)
df_IPL[numeric_cols] = df_IPL[numeric_cols].fillna(0)
```

6) Top 10 Batsmen by Strike Rate


```python
top_batsmen = df_IPL.nlargest(10, "BattingS/R")[["Name", "BattingS/R"]]
plt.figure(figsize=(10, 5))
sns.barplot(y=top_batsmen["Name"], x=top_batsmen["BattingS/R"], color='red')
plt.title("Top 10 Batsmen by Strike Rate")
plt.xlabel("Strike Rate")
plt.ylabel("Player")
plt.show()
```

7) Most Sixes Hit by Players


```python
# Most Sixes Hit by Players
top_six_hitters = df_IPL.nlargest(10, "6s")[["Name", "6s"]]
plt.figure(figsize=(10, 5))
sns.barplot(y=top_six_hitters["Name"], x=top_six_hitters["6s"], color='#4B0082')
plt.title("Most Sixes Hit by Players")
plt.xlabel("Total Sixes")
plt.ylabel("Player")
plt.show()
```

8) Dot Ball Percentage by Bowler


```python
# Dot Ball Percentage by Bowler
df_IPL["Dot Ball %"] = ((df_IPL["Maidens"] * 6) / (df_IPL["Overs"] * 6)) * 100
dot_ball_bowlers = df_IPL.nlargest(10, "Dot Ball %")[["Name", "Dot Ball %"]]
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(y=dot_ball_bowlers["Name"], x=dot_ball_bowlers["Dot Ball %"], color = 'skyblue')
plt.title("Top 10 Bowlers by Dot Ball Percentage")
plt.xlabel("Dot Ball Percentage")
plt.ylabel("Player")
plt.show()
```

13) Top 10 Most Economical Bowlers

```python
economical_bowlers = df_IPL[df_IPL["Overs"] > 20].nsmallest(10, "EconomyRate")[["Name", "EconomyRate"]]
economical_bowlers
```

```python
# Top 10 Most Economical Bowlers
plt.figure(figsize=(10, 5))
sns.barplot(y=economical_bowlers["Name"], x=economical_bowlers["EconomyRate"], color="orange")
plt.title("Top 10 Most Economical Bowlers")
plt.xlabel("Economy Rate")
plt.ylabel("Bowler")
plt.show()
```

```python
df_deliveries = pd.read_csv('/content/drive/MyDrive/Project of Infosys/deliveries.csv')
df_deliveries
```

1) Total Runs Scored Per Over


```python
# Group by 'over' and calculate the sum of 'total_runs'
runs_per_over = df_deliveries.groupby('over')['total_runs'].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='over', y='total_runs', data=runs_per_over, color='skyblue')
plt.title('Total Runs Scored Per Over')
plt.xlabel('Over')
plt.ylabel('Total Runs')
plt.show()
```

3) Total Extra Runs Conceded by Teams


```python
# Group by 'bowling_team' and calculate the sum of 'extra_runs'
extra_runs_by_team = df_deliveries.groupby('bowling_team')['extra_runs'].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='bowling_team', y='extra_runs', data=extra_runs_by_team, color='green')
plt.title('Total Extra Runs Conceded by Teams')
plt.xlabel('Bowling Team')
plt.ylabel('Total Extra Runs')
plt.xticks(rotation=90, ha='right')
plt.show()
```

4) Average Runs Per Over


```python
# Group by 'over' and calculate the mean of 'total_runs'
avg_runs_per_over = df_deliveries.groupby('over')['total_runs'].mean().reset_index()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='over', y='total_runs', data=avg_runs_per_over, color='purple')
plt.title('Average Runs Per Over')
plt.xlabel('Over')
plt.ylabel('Average Runs')
plt.show()
```

5) Distribution of Runs Per Ball (1s, 2s, 4s, 6s)


```python
# Distribution of Runs Per Ball (1s, 2s, 4s, 6s)
runs_per_ball = df_deliveries.groupby('total_runs')['total_runs'].count().reset_index(name='count')
plt.figure(figsize=(8, 6))
sns.barplot(x='total_runs', y='count', data=runs_per_ball[runs_per_ball['total_runs'].isin([1, 2, 4, 6])], color = "gold")
plt.title('Distribution of Runs Per Ball (1s, 2s, 4s, 6s)')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

9) Top 10 Bowlers Conceding Most Boundaries

```python
# Top 10 Bowlers Conceding Most Boundaries
boundaries_conceded = df_deliveries[df_deliveries['batsman_runs'].isin([4, 6])].groupby('bowler')['batsman_runs'].count().reset_index(name='boundaries')
top_boundary_bowlers = boundaries_conceded.nlargest(10, 'boundaries')
plt.figure(figsize=(10, 6))
sns.barplot(x='bowler', y='boundaries', data=top_boundary_bowlers, palette='viridis')
plt.title('Top 10 Bowlers Conceding Most Boundaries')
plt.xlabel('Bowler')
plt.ylabel('Number of Boundaries')
plt.xticks(rotation=45, ha='right')
plt.show()
```

10) Wicket Distribution Across Overs


```python
# Group by 'over' and count the number of dismissals
wickets_per_over = df_deliveries[df_deliveries['player_dismissed'].notna()].groupby('over')['player_dismissed'].count().reset_index(name='wickets')

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='over', y='wickets', data=wickets_per_over, color='red')
plt.title('Wicket Distribution Across Overs')
plt.xlabel('Over')
plt.ylabel('Number of Wickets')
plt.show()
```

11) Total Runs Scored in Powerplay


```python
# Filter the DataFrame to include only the first 6 overs
powerplay_runs = df_deliveries[df_deliveries['over'] <= 6] \
    .groupby('match_id', as_index=False)['total_runs'].sum()

# Create a histogram to visualize the distribution of runs scored in the powerplay
plt.figure(figsize=(12, 6))
sns.histplot(powerplay_runs['total_runs'], bins='auto', color='yellow')

# Improve visualization
plt.title('Distribution of Total Runs Scored in Powerplay', fontsize=14)
plt.xlabel('Total Runs in Powerplay', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
